```python
import pandas as pd
import numpy as np
import itertools
import matplotlib
import matplotlib.pyplot as plt
```

```python
state_feature_vectors = np.asarray(
            list(itertools.product(np.arange(2), repeat=1)))
```

```python
state_feature_vectors
```

```python
states = np.array([np.arange(0, 1, 0.001)])
```

```python
states.shape
```

```python
phi_state = np.cos(np.pi*state_feature_vectors.dot(states))
phi_state.shape
```

```python
for i in range(100):
    theta = pd.read_csv(str(i) + '.csv', header=None).to_numpy()
#     theta = pd.read_csv('91.csv', header=None).to_numpy()
#     print(theta)
#     theta = np.array([[0.01, -0.01, 1, 1]])

    # phi_state = np.transpose(phi_state, (1, 0, 2))
    theta_state = theta.reshape(2, -1).dot(phi_state)

    theta_state = theta_state.T

    theta_state -= np.max(theta_state, axis=1, keepdims=True)
    theta_state = np.exp(theta_state)
    theta_state /= np.sum(theta_state, axis=1, keepdims=True)

#     print(theta_state[:, 0])
    plt.plot(theta_state[:, 0])
```
